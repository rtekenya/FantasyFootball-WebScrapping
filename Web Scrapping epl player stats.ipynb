## Import all the neccessary libraries 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Request the website we will be scrapping and running it with BeautifulSoup

```python
url = 'https://www.fantrax.com/newui/EPL/statsPlayers.go?isSubmit=y&sId=&sortOrder=SCORING_CATEGORY&sortScId=&prevPageNumber=0&pageNumber=&season=921&maxResultsPerPage=50&scKind=NON_GOALIE_STANDARD&position=-2'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
```

## Getting the table

```python
table = soup.find('table', {'class': 'sportsTable curve0022'})
```

```python
headers = []
for i in table.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
df = pd.DataFrame(columns=headers)
```

```python
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text for td in data]
    lenght = len(df)
    df.loc[lenght] = row_data
```

```python
df.to_csv('fantasyEPL2021.csv', index=False)
```
